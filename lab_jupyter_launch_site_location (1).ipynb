```python
from math import sin, cos, sqrt, atan2, radians
import folium
from folium.features import DivIcon

# Define distance calculation (Haversine formula)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # radius of Earth in km

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    
    distance = R * c
    return distance

# Example launch site coordinate (replace with your actual launch site lat/long)
launch_site_lat = 28.5623
launch_site_lon = -80.5774
launch_site_coord = [launch_site_lat, launch_site_lon]

# Example proximity coordinates (replace with actual coordinates you get via MousePosition)
coastline_lat = 28.56367
coastline_lon = -80.57163
coastline_coord = [coastline_lat, coastline_lon]

# Calculate distance
distance_to_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance to coastline: {distance_to_coastline:.2f} km")

# Add marker for coastline point with distance label
distance_marker = folium.Marker(
    location=coastline_coord,
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12pt; color:#d35400;"><b>{distance_to_coastline:.2f} KM</b></div>'
    )
)

# Draw line between launch site and coastline
line = folium.PolyLine(locations=[launch_site_coord, coastline_coord], weight=2, color='blue')

# Add launch site marker
launch_site_marker = folium.Marker(
    location=launch_site_coord,
    popup='Launch Site',
    icon=folium.Icon(color='red', icon='rocket', prefix='fa')
)

# Initialize the map centered at launch site
site_map = folium.Map(location=launch_site_coord, zoom_start=12)

# Add markers and line to map
site_map.add_child(launch_site_marker)
site_map.add_child(distance_marker)
site_map.add_child(line)

# Display the map
site_map

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
from math import radians, cos, sin, asin, sqrt

def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the Haversine distance between two lat/lon coordinates in kilometers.
    """
    R = 6371  # Radius of Earth in kilometers

    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine calculation
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    distance = R * c

    return distance

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Coordinates
launch_site = (28.562302, -80.577356)
city = (28.3922, -80.6077)

# Distance calculation
def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

distance_km = haversine(*launch_site, *city)

# Plot
plt.figure(figsize=(8, 6))
plt.plot([launch_site[1], city[1]], [launch_site[0], city[0]], 'b--', label='Distance Line')
plt.scatter(*launch_site[::-1], color='green', s=100, label='Launch Site')
plt.scatter(*city[::-1], color='red', s=100, label='Closest City')

plt.text(city[1], city[0], f'{distance_km:.2f} KM', fontsize=12, color='darkorange')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Launch Site and Closest City')
plt.legend()
plt.grid(True)
plt.show()


```
